```python
import os
from ragflow_sdk import RAGFlow

# RAGFlow 설정
API_KEY = "ragflow-I0Y2ViN2M2MDU3OTExZjA5N2ZhMTIxOD"  # 실제 RAGFlow API 키로 교체
BASE_URL = "http://localhost"  # RAGFlow 서버 URL
rag_object = RAGFlow(api_key=API_KEY, base_url=BASE_URL)

# 데이터셋 가져오기
def get_datasets(rag):
    definition = rag.get_dataset(name="Definition")
    paper = rag.get_dataset(name="Paper")
    return definition, paper

# 문서 목록 출력
def print_document_list(dataset, dataset_name):
    print(f"\n=== Documents in {dataset_name} ===")
    for doc in dataset.list_documents():
        print(f"- {doc.name}")

# Definition 데이터셋의 청크 출력
def print_definition_chunks(definition):
    print("\n=== Chunks in Definition Dataset ===")
    for doc in definition.list_documents():
        print(f"\nDocument: {doc.name}")
        for chunk in doc.list_chunks():
            print(f"Chunk: {chunk.content[:100]}...")  # 처음 100자만 출력

# Paper 데이터셋에서 파일별 답변 생성
def get_file_based_response(rag, dataset, prompt):
    responses = {}
    for doc in dataset.list_documents():
        # 각 문서에 대해 개별적으로 챗봇 생성
        assistant = rag.create_chat(f"Chat_{doc.name}", dataset_ids=[dataset.id])
        session = assistant.create_session(f"Session for {doc.name}")
        
        # 문서의 청크를 기반으로 프롬프트에 답변 생성
        full_content = ""
        for chunk in doc.list_chunks():
            full_content += chunk.content + "\n"
        
        # 프롬프트에 파일 이름을 명시적으로 포함
        file_prompt = f"{prompt} (Based on file: {doc.name})"
        response = ""
        for ans in session.ask(file_prompt, stream=True):
            response += ans.content
        
        responses[doc.name] = response
    return responses

# 대화형 인터페이스
def interactive_chat(rag, paper_dataset):
    # Paper 데이터셋의 모든 ID 수집
    datasets = rag.list_datasets(name="Paper")
    dataset_ids = [dataset.id for dataset in datasets]
    
    # 기본 챗봇 생성
    assistant = rag.create_chat("Miss RRR", dataset_ids=dataset_ids)
    initial_prompt = "What is the main topic of these PDFs?"
    session = assistant.create_session(initial_prompt)

    print("\n==================== Miss RR ====================\n")
    print("Hello. What can I do for you?")
    
    while True:
        question = input("\n==================== User ====================\n> ")
        if question.lower() in ["exit", "quit"]:
            print("\n==================== Miss RR ====================\n")
            print("Goodbye.")
            break
        
        print("\n==================== Miss RR ====================\n")
        print(f"Processing question: '{question}'")
        
        # 파일별 답변 생성
        file_responses = get_file_based_response(rag, paper_dataset, question)
        
        # 파일별로 답변 출력
        for file_name, response in file_responses.items():
            print(f"\n--- Response for {file_name} ---")
            print(response)
        
        # 전체 데이터셋에 대한 일반 답변도 추가 (옵션)
        general_response = ""
        for ans in session.ask(question, stream=True):
            print(ans.content, end='', flush=True)
            general_response += ans.content
        print("\n")

# 메인 실행
def main():
    # 데이터셋 가져오기
    definition, paper = get_datasets(rag_object)
    
    # 문서 목록 출력
    print_document_list(definition, "Definition")
    print_document_list(paper, "Paper")
    
    # Definition 청크 출력
    print_definition_chunks(definition)
    
    # 대화형 챗봇 실행
    interactive_chat(rag_object, paper)

if __name__ == "__main__":
    main()
```

```python
import os
from ragflow_sdk import RAGFlow
from ragflow_sdk.modules.chat import Chat
from datetime import datetime
import json
import pandas as pd
from typing import Dict, List, Tuple
import re
import yaml

class RagflowAnalyzer:
    """
    RAGFlow를 이용한 보고서 분석기
    
    Attributes:
        api_key (str): RAGFlow API 키
        base_url (str): RAGFlow 서버 URL
        rag_object (RAGFlow): RAGFlow 객체
        definition_dataset: Definition 데이터셋
        paper_dataset: Paper 데이터셋
        sector_data (dict): 섹터 데이터
        event_data (dict): 이벤트 데이터
        chat_assistant: 챗봇 인스턴스 저장
    """
    
    def __init__(self, api_key: str, base_url: str = "http://localhost"):
        """
        RagflowAnalyzer 초기화
        
        Args:
            api_key (str): RAGFlow API 키
            base_url (str): RAGFlow 서버 URL
        """
        self.api_key = api_key
        self.base_url = base_url
        self.rag_object = RAGFlow(api_key=api_key, base_url=base_url)
        self.definition_dataset = self.rag_object.get_dataset(name="Definition")
        self.paper_dataset = self.rag_object.get_dataset(name="Paper")
        self.sector_data = self._load_sector_data()
        self.event_data = self._load_event_data()
        self.chat_assistant = None  # 챗봇 인스턴스 저장
        
    def _load_sector_data(self) -> dict:
        """섹터 데이터를 로드합니다."""
        for doc in self.definition_dataset.list_documents():
            if doc.name == "sector.json":
                for chunk in doc.list_chunks():
                    return json.loads(chunk.content)
        return {}
    
    def _load_event_data(self) -> dict:
        """이벤트 데이터를 로드합니다."""
        for doc in self.definition_dataset.list_documents():
            if doc.name == "event.json":
                for chunk in doc.list_chunks():
                    return json.loads(chunk.content)
        return {}
    
    def parse_filename(self, filename: str) -> Dict[str, str]:
        """
        파일명을 파싱하여 종목명, 작성기관, 작성일을 추출합니다.
        
        Args:
            filename (str): 파싱할 파일명
            
        Returns:
            Dict[str, str]: 파싱된 정보를 담은 딕셔너리
        """
        # 파일 확장자 제거
        name = os.path.splitext(filename)[0]
        print(f"\n=== 파일명 파싱 디버깅 ===")
        print(f"원본 파일명: {filename}")
        print(f"확장자 제거: {name}")
        
        # 언더스코어로 분리
        parts = name.split('_')
        print(f"분리된 부분: {parts}")
        
        # 마지막 4개 요소 추출
        date_str = parts[-1]  # 25.02.18
        institution = parts[-3]  # 미래에셋증권
        company = parts[-4]  # HD현대중공업
        
        print(f"추출된 정보:")
        print(f"- 날짜: {date_str}")
        print(f"- 기관: {institution}")
        print(f"- 회사: {company}")
        
        # 날짜 형식 변환 (25.02.18 -> 2025-02-18)
        date_obj = datetime.strptime(date_str, '%y.%m.%d')
        formatted_date = date_obj.strftime('%Y-%m-%d')
        
        return {
            'company': company,
            'institution': institution,
            'date': formatted_date,
            'original_filename': filename
        }
    
    def get_sorted_documents(self) -> List[Dict[str, str]]:
        """
        데이터셋의 문서들을 파싱하고 날짜순으로 정렬합니다.
        
        Returns:
            List[Dict[str, str]]: 정렬된 문서 목록
        """
        documents = []
        for doc in self.paper_dataset.list_documents():
            parsed_info = self.parse_filename(doc.name)
            documents.append(parsed_info)
        return sorted(documents, key=lambda x: x['date'], reverse=True)
    
    def _get_document_chunks(self, doc_name: str) -> List[str]:
        """
        문서의 청크들을 검색합니다.
        
        Args:
            doc_name (str): 문서 이름
            
        Returns:
            List[str]: 문서 청크 내용 리스트
        """
        chunks = []
        for doc in self.paper_dataset.list_documents():
            if doc.name == doc_name:
                for chunk in doc.list_chunks():
                    chunks.append(chunk.content)
        return chunks

    def _get_or_create_chat_assistant(self) -> 'Chat':
        """
        챗봇을 가져오거나 생성합니다.
        이미 존재하는 경우 기존 챗봇을 반환하고, 없는 경우 새로 생성합니다.
        
        Returns:
            Chat: 챗봇 인스턴스
        """
        CHAT_NAME = "Report_Analyzer"
        
        if self.chat_assistant is None:
            try:
                # 챗봇 생성 시도
                self.chat_assistant = self.rag_object.create_chat(
                    name=CHAT_NAME,
                    dataset_ids=[self.paper_dataset.id]
                )
                
                # 챗봇 설정 업데이트
                self.chat_assistant.update({
                    "llm": {
                        "model_name": "gpt-4o-mini",
                        "temperature": 0.1,
                        "top_p": 0.3,
                        "presence_penalty": 0.2,
                        "frequency_penalty": 0.7,
                        "max_tokens": 4096
                    },
                    "prompt": {
                        "similarity_threshold": 0.2,
                        "keywords_similarity_weight": 0.7,
                        "top_n": 8,
                        "show_quote": True,  # 인용 표시 활성화
                        "variables": [
                            {"key": "company", "optional": False},
                            {"key": "institution", "optional": False},
                            {"key": "date", "optional": False},
                            {"key": "filename", "optional": False},
                            {"key": "content", "optional": False}  # 문서 내용 변수 추가
                        ],
                        "prompt": (
                            "당신은 금융 보고서 분석 전문가입니다. "
                            "다음 보고서를 분석하여 주요 내용을 정리해주세요:\n\n"
                            "문서명: {filename}\n"
                            "회사: {company}\n"
                            "작성기관: {institution}\n"
                            "작성일: {date}\n\n"
                            "분석할 문서 내용:\n{content}\n\n"
                            "다음 형식으로 YAML 응답해주세요:\n"
                            "investment_point:\n"
                            "  key_points:\n"
                            "    - 포인트1\n"
                            "    - 포인트2\n"
                            "    - 포인트3\n"
                            "  reasonings:\n"
                            "    - 근거1\n"
                            "    - 근거2\n"
                            "    - 근거3\n"
                            "analysis:\n"
                            "  figure:\n"
                            "    fact:\n"
                            "      - 사실1\n"
                            "      - 사실2\n"
                            "    opinion:\n"
                            "      - 의견1\n"
                            "      - 의견2\n"
                            "  nonfigure:\n"
                            "    fact:\n"
                            "      - 사실1\n"
                            "      - 사실2\n"
                            "    opinion:\n"
                            "      - 의견1\n"
                            "      - 의견2\n"
                            "  material:\n"
                            "    fact:\n"
                            "      - 사실1\n"
                            "      - 사실2\n"
                            "    opinion:\n"
                            "      - 의견1\n"
                            "      - 의견2\n"
                            "  public:\n"
                            "    fact:\n"
                            "      - 사실1\n"
                            "      - 사실2\n"
                            "    opinion:\n"
                            "      - 의견1\n"
                            "      - 의견2\n"
                            "sector_indicators:\n"
                            "  - 지표1\n"
                            "  - 지표2\n"
                            "events:\n"
                            "  - category: 이벤트 카테고리\n"
                            "    type: 이벤트 타입\n"
                            "    description: 이벤트 설명\n"
                            "    probability: 높음/중간/낮음\n\n"
                            "다른 설명이나 텍스트는 포함하지 마세요."
                        )
                    }
                })
                print(f"새로운 챗봇 생성됨: {CHAT_NAME}")
                
            except Exception as e:
                if "Duplicated chat name" in str(e):
                    # 이미 존재하는 경우 기존 챗봇 찾기
                    chats = self.rag_object.list_chats(name=CHAT_NAME)
                    if chats:
                        self.chat_assistant = chats[0]
                        print(f"기존 챗봇 사용: {CHAT_NAME}")
                    else:
                        raise Exception(f"챗봇을 찾을 수 없음: {CHAT_NAME}")
                else:
                    raise e
                    
        return self.chat_assistant
    
    def _extract_json_from_response(self, response: str) -> dict:
        """
        응답에서 YAML 형식의 데이터를 추출하여 JSON으로 변환합니다.
        
        Args:
            response (str): 챗봇 응답
            
        Returns:
            dict: 추출된 데이터
        """
        try:
            # 응답에서 YAML 부분만 추출
            yaml_str = response.strip()
            if yaml_str.startswith('**ERROR**'):
                print(f"\n=== 오류 응답 ===")
                print(yaml_str)
                return {}
            
            # YAML 시작 부분 찾기
            yaml_start = yaml_str.find("investment_point:")
            if yaml_start == -1:
                print("\n=== YAML 시작 부분을 찾을 수 없음 ===")
                print("전체 응답:")
                print(yaml_str)
                return {}
                
            # YAML 부분만 추출
            yaml_content = yaml_str[yaml_start:]
            
            # YAML 파싱 시도
            try:
                result = yaml.safe_load(yaml_content)
                if not result:
                    print("\n=== YAML 파싱 결과가 비어있음 ===")
                    print("YAML 내용:")
                    print(yaml_content)
                    return {}
                    
                print("\n=== 추출된 데이터 ===")
                print(json.dumps(result, ensure_ascii=False, indent=2))
                return result
                
            except yaml.YAMLError as e:
                print(f"\n=== YAML 파싱 오류 ===")
                print(f"오류 메시지: {str(e)}")
                print("YAML 내용:")
                print(yaml_content)
                return {}
            
        except Exception as e:
            print(f"\n=== 처리 중 오류 발생 ===")
            print(f"오류 메시지: {str(e)}")
            print("전체 응답:")
            print(response)
            return {}
    
    def analyze_report(self, doc_name: str) -> Dict:
        """
        개별 보고서를 분석합니다.
        
        Args:
            doc_name (str): 분석할 문서 이름
            
        Returns:
            Dict: 분석 결과
        """
        # 문서 정보 파싱
        doc_info = self.parse_filename(doc_name)
        print(f"\n=== 문서 정보 ===")
        print(f"파일명: {doc_name}")
        print(f"회사명: {doc_info['company']}")
        print(f"작성기관: {doc_info['institution']}")
        print(f"작성일: {doc_info['date']}")
        
        # 문서 청크 검색
        chunks = self._get_document_chunks(doc_name)
        if not chunks:
            print(f"문서 청크를 찾을 수 없음: {doc_name}")
            return {
                **doc_info,
                "analysis": {},
                "error": "문서 청크를 찾을 수 없음"
            }
            
        # 문서 내용 결합
        content = "\n\n".join(chunks)
        print(f"\n=== 문서 청크 수: {len(chunks)} ===")
        
        # 챗봇 가져오기 또는 생성
        assistant = self._get_or_create_chat_assistant()
        
        # 세션 생성
        session = assistant.create_session(f"Analysis for {doc_name}")
        
        try:
            print("\n=== 챗봇 응답 처리 시작 ===")
            
            # 분석 실행 (스트리밍 활성화)
            response = session.ask(
                question="",  # 빈 질문으로 프롬프트의 변수만 사용
                stream=False,  # 스트리밍 활성화
                company=doc_info['company'],
                institution=doc_info['institution'],
                date=doc_info['date'],
                filename=doc_name,
                content=content  # 문서 내용 전달
            )
            
            print("*"*20, "\n챗봇 응답", "*"*20)
            print(response)
            print("*"*20, "\n챗봇 응답", "*"*20)
            
            # 결과 통합
            analysis_result = self._extract_json_from_response(response)
            if not analysis_result:
                print("\n=== 분석 결과가 비어있음 ===")
                return {
                    **doc_info,
                    "analysis": {},
                    "error": "분석 결과 추출 실패"
                }
                
            result = {
                **doc_info,
                "analysis": analysis_result
            }
            
            # 결과 저장 시도
            try:
                self.save_to_yaml([result], "test_result.yaml")
                self.save_to_excel([result], "test_result.xlsx")
                print("\n=== 결과 저장 완료 ===")
            except Exception as e:
                print(f"\n=== 결과 저장 중 오류 발생 ===")
                print(str(e))
            
            return result
            
        except Exception as e:
            print(f"\n=== 분석 중 오류 발생 ===")
            print(str(e))
            return {
                **doc_info,
                "analysis": {},
                "error": str(e)
            }
    
    def analyze_all_reports(self) -> List[Dict]:
        """
        모든 보고서를 분석합니다.
        
        Returns:
            List[Dict]: 분석된 모든 보고서 결과
        """
        results = []
        for doc in self.get_sorted_documents():
            result = self.analyze_report(doc['original_filename'])
            results.append(result)
        return results
    
    def save_to_yaml(self, results: List[Dict], output_file: str = "analysis_results.yaml"):
        """
        분석 결과를 YAML 파일로 저장합니다.
        
        Args:
            results (List[Dict]): 저장할 분석 결과
            output_file (str): 출력 파일 경로
        """
        with open(output_file, 'w', encoding='utf-8') as f:
            yaml.dump(results, f, allow_unicode=True, sort_keys=False, indent=2)
            
    def save_to_excel(self, results: List[Dict], output_file: str = "analysis_results.xlsx"):
        """
        분석 결과를 Excel 파일로 저장합니다.
        
        Args:
            results (List[Dict]): 저장할 분석 결과
            output_file (str): 출력 파일 경로
        """
        # 데이터 정리
        excel_data = []
        for result in results:
            # 기본 정보
            row = {
                '회사명': result['company'],
                '작성기관': result['institution'],
                '작성일': result['date'],
                '파일명': result['original_filename']
            }
            
            # 분석 결과
            analysis = result.get('analysis', {})
            
            # 투자 포인트
            investment_point = analysis.get('investment_point', {})
            row['주요 포인트'] = '\n'.join(investment_point.get('key_points', []))
            row['근거'] = '\n'.join(investment_point.get('reasonings', []))
            
            # 분석 결과
            for category in ['figure', 'nonfigure', 'material', 'public']:
                cat_data = analysis.get(category, {})
                row[f'{category}_사실'] = '\n'.join(cat_data.get('fact', []))
                row[f'{category}_의견'] = '\n'.join(cat_data.get('opinion', []))
            
            # 섹터 지표
            row['섹터 지표'] = '\n'.join(analysis.get('sector_indicators', []))
            
            # 이벤트
            events = analysis.get('events', [])
            for i, event in enumerate(events, 1):
                row[f'이벤트{i}_카테고리'] = event.get('category', '')
                row[f'이벤트{i}_타입'] = event.get('type', '')
                row[f'이벤트{i}_설명'] = event.get('description', '')
                row[f'이벤트{i}_확률'] = event.get('probability', '')
            
            excel_data.append(row)
        
        # 데이터프레임 생성 및 저장
        df = pd.DataFrame(excel_data)
        df.to_excel(output_file, index=False)
```
